```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import SGDRegressor
from sklearn.preprocessing import StandardScaler
```

```python
# Dataset
X_train = [[3],[3.5],[4],[4.5],[5],[5.5],[6],[6.5],[7]]
Y_train = [[3.57],[4.9],[5.81],[6.51],[7.18],[8.01],[8.62],[9.05],[9.54]]
X_test = [[3.2],[3.75],[3.89],[4.2],[4.69],[5.1],[6.9],[7.7],[8.2]]
```

```python
scale = StandardScaler() # Instance for StandardScaler
X_norm = scale.fit_transform(X_train)  # fit and then standardize
X_test_norm = scale.fit_transform(X_test)  # Normalizing test set
print(X_norm)
```

```python
sgdr = SGDRegressor(max_iter = 1000) #number of epochs
'''This is an instance for Stochastic Gradient Descent. It is not 
started to run yet.'''

sgdr.fit(X_norm, Y_train)   # Run the SGD Regression for 1000 epochs
print("Number of iteraions : ", sgdr.n_iter_)   # gives 179, because it converge to the lowest point of loss function at 179 epochs. 
```

```python
# Weights and Bias terms
b_norm = sgdr.intercept_
w_norm = sgdr.coef_
print("w_norm : ", w_norm)
print("b_norm : ", b_norm)
```

```python
# Prediction 
y_pred = sgdr.predict(X_norm)   # Prediction for training dataset
y_pred_test = sgdr.predict(X_test_norm)   # Prediction for test dataset
```

```python
# Plot results
fig,ax=plt.subplots(1,1,figsize=(8,5),sharey=True)

ax.scatter(X_train,Y_train, label = 'target')
ax.scatter(X_train,y_pred, label = 'predictions')
#ax.scatter(X_test,y_pred_test, label = 'test predictions')
ax.legend()
ax.grid(True)
plt.show()
```
